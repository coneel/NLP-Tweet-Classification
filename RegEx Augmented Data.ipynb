```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

import re
import random
```

```python
df = pd.read_csv('/Users/caleboneel/Downloads/train.csv')
```

```python
df['link'] = 0
```

```python
text = df['text']
```

```python
url = 'https?:\/\/.*'
link_df = df.copy()
for index, tweet in enumerate(text):
    words = tweet.split()
    for w_index, w in reversed(list(enumerate(words))):
        if re.search(url,w):
            link_df.loc[index, 'link'] = 1
        else:
            pass
link_df['text'] = link_df['text'].str.replace('https?:\/\/.*', '')
```
